```r
# Select any external online data source (It could be from your drive, single file data set will be ok)
# not native to R internal data packages. Write R code to pull that online data set into your colab
# notebook. While selecting data set it should have atleast 4 attributes and more than 100 sample data
# points.
data <- read.csv("data.csv")
data
```

```r
# a. print the 5 point statistical summary for any 2 numeric attributes from your data set
fivenum(data$area_mean) 
fivenum(data$radius_mean)
```

```r
# b. select 2 atributes have closest mean value and draw a scatter plot with proper x and labelslointt and title 
# (for plotting you must use R library functions or gnuplot and not matplotlib )
means<- colMeans(data[3:7])
means
```

```r
plot(data$radius_mean, data$texture_mean, main='radius_mean', xlab='radius_mean', ylab='texture_mean', xlim=c(0,25), ylim=c(0,25))
```

```r
# c. Find coefficient of correlation using some R library function between above 2 attributes
cor(data$radius_mean,data$texture_mean,method ="pearson")
cor(data$radius_mean,data$texture_mean,method ="kendall")
cor(data$radius_mean,data$texture_mean,method ="spearman")
```

```r
# d. Draw a box plot for each of the two attributes selected above in a single plot
bp <- data.frame(values=c(data$radius_mean, data$texture_mean), group=c(rep('radius_mean', length(data$radius_mean)), group=rep('texture_mean', length(data$texture_mean))))
boxplot(values ~ group, bp)
```
